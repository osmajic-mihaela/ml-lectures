# **Šta je Mašinsko učenje?**

---



###  🌟 Analogija
Zamislite da želite da naučite prijatelja da prepozna voće u prodavnici. Umesto da mu date knjigu sa pravilima "jagode su crvene, banane su žute" (što bi predstavljalo tradicionalno programiranje korišćenjem pravila), vi mu pokažete 100 slika različitog voća i kažete: "Ovo je jabuka, ovo je banana, ovo je ananas...". Nakon toga, prijatelj će sam početi da primećuje obrasce (npr. "banane su duguljaste i žute") i moći će da prepozna voće sa nove slike, koju nikad pre nije video.

Mašinsko učenje radi isto – uči iz primera da donese odluke.

---

### 📚  Definicija
**Mašinsko učenje je način da računari nauče da rešavaju zadatke bez eksplicitnog programiranja. Umesto da im kažemo tačno šta da rade, mi im damo podatke i pustimo ih da pronađu obrasce.**

---

### Pojmovi:  
- **Podaci**
- **Učenje**: Računar traži obrasce u podacima
- **Predviđanje**: Kad nauči obrasce, može da odgovori na nova pitanja

# **Koje sve vrste mašinskog učenja postoje?**

---



### 🌟 Analogija
Zamislite tri načina na koja dete uči da prepoznaje životinje:
1. **Sa učiteljem**: Učitelj pokazuje detetu sliku po sliku i govori šta je na njoj ("ovo je lav", "ovo je slon"), dete uči i vežba da ih prepozna uz korekcije od strane učitelja.  
2. **Bez učitelja**: Dete dobija gomilu slika bez natpisa i grupše ih prema sličnostima (npr. prema boji).  
3. **Kroz igru**: Dete pokušava da pronađe igračku životinje, a roditelji ga navode instrukcijama "toplo" kad ide u dobrom smeru ili "hladno" kad greši.  

**Ovo su tri tipa mašinskog učenja!**

---

### 📚 Definicija
Mašinsko učenje se deli na tri glavna tipa, zavisno od toga **kako model uči**:  



### 🏷️ **1. Nadgledano učenje (Supervised Learning)**  
-  Model uči iz **označenih podataka** – ima "odgovor" za svaki primer.  
- **Analogija**: Flashcard učenje. (Flashcard je kartica za učenje koja pomaže u vežbanju pamćenja, pitanje se nalazi na jednoj strani kartice, a odgovor na drugoj)
- **Primeri**:  
  - Klasifikacija: Razvrstavanje mejlova u "spam" ili "ne-spam" kategorije, na osnovu uočenih šablona postojećih mailova označenih kao "spam" ili "ne-spam".
  - Regresija: Predviđanje cene stanova na osnovu kvadrature i lokacije, na osnovu uočenih šablona .  
- **Ključna reč**: **"Znamo oznake - tačne odgovore."**  



### 🧩 **2. Nenadgledano učenje (Unsupervised Learning)**  
- Model traži obrasce u **neoznačenim podacima** – nema "odgovora".  
- **Analogija**: Dete sortira igračke u grupe, a da ne zna njihova imena.  
- **Primeri**:  
  - Klasterovanje: Grupisanje korisnika po sličnim navikama kupovine.  
  - Smanjivanje dimenzija: Prepoznati dominantne karakteristike kompleksnih podataka i pojednostaviti ih (npr. za vizuelizaciju). Detaljnije u narednim lekcijama.
- **Ključna reč**: **"Nemamo oznake - tražimo skrivene obrasce."**  


### 🎮 **3. Učenje sa pojačanjem (Reinforcement Learning)**  
-  Model uči **kroz pokušaje i greške**, primajući nagrade/kazne.  
- **Analogija**: Učenje psa trikovima – dobija poslasticu za dobru akciju, ignorisanje za lošu.  
- **Primeri**:  
  - Igre: AlphaGo (program je sam učio da igra i kroz poraze i pobede naučio da igra tako da pobedi čoveka) .  
  - Robotika: Robot koji uči da hoda bez pada, kroz samounapređivanje.  
- **Ključna reč**: **"Uči sa nagradama i kaznama"**

# **Kako izgleda radni tok u ML projektu?**

---



### 🌟 **Analogija: ML projekat je kao kuvanje jelа**  
Da biste napravili jelo, morate proći kroz faze: planiranje, sakupljanje sastojaka, pripremu, kuvanje, probu i serviranje.

 ML projekti rade na sličan način – samo umesto hrane, pravite model koji rešava problem!

U nastavku ćemo povući paralele između koraka kuvanja jela i ML projekta, a u sledećoj lekciji svaki korak detaljnije opisati i objasniti.

---

![img/1/ml-workflow.jpg](img/1/ml-workflow.jpg)

---

### 📋 **Korak 1: Definišite problem i cilj**  
- Određujete **šta ćete kuvati**.   
- **Primer**: Hoćemo da predviđamo cenu stanova u Beogradu na osnovu kvadrature, lokacije i godišta.

---

### 🛒 **Korak 2: Prikupite i istražite podatke**  
- **Kupovina sastojaka**.  
- **Akcije**:  
  -  Prikupite podatke iz baza podataka, tekstualnih fajlova, API-ja.  
  - Istražite podatke: Da li nedostaju vrednosti? Ima li izuzetaka u vrednostima, koji značajno odstupaju od tipičnog uticaja kvadrature, lokacije i godišta na cenu (eng. outliers)?  

  ```  
  Podaci :
  | Kvadratura  |   Lokacija  | Godište |   Cena  |  
  |-------------|-------------|---------|---------|
  |     58      | Vračar      |   1990  | 200.000 |  
  |     65      | Zvezdara    |   2005  | 210.000 |  
  |     ?       | Karaburma   |   1985  | 198.000 |  (nedostaje kvadratura)  
  |     59      | Dorćol      |   1992  | 850.000 |  (outlier)
  ```  

---

### 🧹 **Korak 3: Očistite i pripremite podatke (Preprocessing)**  
- **Sečenje i pranje sastojaka** pre kuvanja.  
- **Akcije**:  
  - Popunite nedostajuće vrednosti (npr. prosečna kvadratura primera sa sličnim vrednostima atributa).  
  - Kodirajte kategorijske promenljive (npr. "Novi Sad" → [0,1], "Novi Beograd" → [1,0]). Više o ovome u nastavku.
  - Normalizujte brojčane vrednosti (npr. sve cene skalirajte između 0 i 1).  
-  
  ```  
  "Vračar" → [0,1] ✔️  
  Nedostajuća kvadratura → 75 (prosek) ✔️  
  Cena od 200.000€ → 0.8 (nakon skaliranja) ✔️  
  ```  

---

### 🧠 **Korak 4: Obučite model**  
- **Kuvanje**   
- **Akcije**:  
  1. Podelite podatke na **trening** (70-80%), **validacioni** (15-10%) i **test** skup (15-10%).  
  2. Odaberite algoritam (npr. linearna regresija, Random Forest).  
  3. Obučite model na trening podacima.  
- **Primer**:  
  Skup podataka se sastoji od mnogo pojedinačnih primera stanova – svaki primer predstavlja jedan stan sa svojim karakteristikama, kao što su kvadratura, broj soba, sprat, cena... Obučavanje znači da računar "gleda" te primere i pokušava da pronađe pravilo, tj. formulu, koja povezuje kvadraturu i druge karakteristike sa cenom. Tu formulu zovemo **model**, a brojevi unutar nje, koji se prilagođavaju (kroz proces obučavanja) da bi formula što bolje radila, zovemo **parametri**.     
  Pored toga, postoji i nekoliko stvari koje sami biramo pre nego što obučavanje počne – to su **hiperparametri**. Oni određuju kako model uči. Nakon obučavanja, model može da proceni cenu stana kog nikad nije video.

---

### 📊 **Korak 5: Evaluacija i fino podešavanje**  
- **Probate jelo i dodajte začine** ako nije savršeno, potom ga servirate za testiranje kada je gotovo.
- **Metrike za procenu performanse**:  
  - Odnos stvarne i predviđene vrednosti od strane modela.
  - Za regresiju: MSE, R².  
  - Za klasifikaciju: Tačnost, F1-score.  
- **Podešavanje**:  
  - Optimizujte hiperparametre na validacionim podacima.  
- **Testiranje**:  
  - Kada su hiperparametri optimizovani, model obučen sa njima, testirajte model na test podacima.  

  ```  
  R² na test skupu = 0.89 → Dobro!  
  MSE = 15.000 → Može bolje.  
  ```    

---

### 🌍 **Primer: Preporuka filmova**  
1. **Cilj**: Preporučiti filmove koje će korisnik voleti.  
2. **Podaci**:

    Istorija gledanja:  Koje filmove je korisnik gledao i koliko često   
    Ocene:  Ocene koje su korisnici dali filmovima  
    Žanrovi:  Kategorije filmova    
    Metapodaci o filmu: Godina izdavanja, reditelj, glumci, trajanje

3. **Čišćenje**: Uklanjanje duplih filmova, popunjavanje nedostajućih ocena...  
4. **Model**: Korišćenje KNN algoritma (Koji filmovi su najsličniji filmovima koji su se svideli korisniku?)
5. **Evaluacija**: Provera da li preporučeni filmovi imaju visoke ocene kod korisnika.
